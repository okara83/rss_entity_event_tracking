```python
import json
from elasticsearch import Elasticsearch
host="https://7771acde721f4a6bafa7cc6819bf7a45.us-east-1.aws.found.io:9243"
creds = json.load(open("es_creds.json"))
username=creds["username"]
pw=creds["password"]
es = Elasticsearch([host], http_auth=(username,pw))
```

```python

res = es.search(
    index="tracking_events_prod",
    body={
        "query": 
        {
            "match_all": {}
        }
    }
)

print(json.dumps(res,indent=5))
```

```python
### the inverse of match_all is match_none (therefore this which matches no documents.)
res = es.search(
    index="tracking_events_prod",
    body={
        "query": 
        {
            "match_none": {}
        }
    }
)

print(json.dumps(res,indent=5))
```

```python
res = es.search(
    index="tracking_events_prod",
    body={
        "query": 
        {
            "match": {
                "entity.keyword": "Donald Trump"
            }
        }
    }
)
print(json.dumps(res,indent=5))
res.keys()
```

```python
## for entity term search, use entity.keyword
res = es.search(
    index="tracking_events_prod", 
    body={
        "query": {
        "bool": {
          "must": [],
          "filter": [
            {
              "bool": {
                "should": [
                  {
                    "match": {
                      "entity.keyword": "Donald Trump"
                    }
                  }
                ],
                "minimum_should_match": 1  ###not necessary
              }
            },
            {
              "range": {
                "event_datetime": {
                  "format": "strict_date_optional_time",
                  "gte": "2019-02-07T17:20:25.832Z",
                  "lte": "2020-02-07T17:20:25.832Z"
                }
              }
            }
          ],
          "should": [],
          "must_not": []
        }
      }
    })

print(json.dumps(res,indent=5))
```

NOTICE THE ABOVE QUERY MATCHES THE ENTITY "Donald Trump" ONLY, WHEREAS THE NEXT QUERY RETURN NOT ONLY "Donald Trump" BUT ALSO "Donald Duck"

```python
### entity instead of entity.keyword
res = es.search(
    index="tracking_events_prod", 
    body={
        "query": {
        "bool": {
          "must": [],
          "filter": [
            {
              "bool": {
                "should": [
                  {
                    "match": {
                      "entity": "Donald Trump"
                    }
                  }
                ],
                "minimum_should_match": 1 ###not necessary
              }
            },
            {
              "range": {
                "event_datetime": {
                  "format": "strict_date_optional_time",
                  "gte": "2019-02-07T17:20:25.832Z",
                  "lte": "2020-02-07T17:20:25.832Z"
                }
              }
            }
          ],
          "should": [],
          "must_not": []
        }
      }
    })

```
